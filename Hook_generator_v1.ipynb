```python
# --- Hook Generator Form (Colab Widgets) ---

import os
from openai import OpenAI
from google.colab import userdata
import ipywidgets as widgets
from IPython.display import display, clear_output

# Ensure API key is set (you already confirmed this works)
os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")
client = OpenAI()

# Widgets (form inputs)
platform_w = widgets.Dropdown(
    options=["TikTok", "Instagram Reels", "YouTube Shorts"],
    value="TikTok",
    description="Platform:"
)

audience_w = widgets.Text(value="South Africa", description="Audience:")
niche_w = widgets.Text(value="podcast clips", description="Niche:")
clip_type_w = widgets.Dropdown(
    options=["argument", "confession", "funny", "lesson", "shocking", "wholesome", "drama"],
    value="argument",
    description="Clip type:"
)

emotion_w = widgets.Text(value="Shock and curiosity", description="Emotion:")
vibe_w = widgets.Dropdown(
    options=["Serious", "Funny", "Calm", "Aggressive", "Luxury", "Street"],
    value="Serious",
    description="Vibe:"
)

notable_individuals_w = widgets.Text(value="", description="Notable Individuals")

length_w = widgets.Dropdown(
    options=["7 seconds", "15 seconds", "30 seconds", "45 seconds"],
    value="15 seconds",
    description="Length:"
)

goal_w = widgets.Dropdown(
    options=["maximize retention", "maximize shares", "maximize comments", "maximize follows", "maximize views"],
    value="maximize retention",
    description="Goal:"
)

forbidden_w = widgets.Text(value="", description="Forbidden:")
refs_w = widgets.Textarea(
    value="This changed how I see everything\nI did not expect him to say this\nThis got awkward fast",
    description="Refs:",
    layout=widgets.Layout(width="700px", height="100px")
)

generate_btn = widgets.Button(description="Generate 10 Hooks", button_style="success")
output = widgets.Output()

def build_prompt():
    ref_lines = [ln.strip() for ln in refs_w.value.split("\n") if ln.strip()]
    forbidden = forbidden_w.value.strip() if forbidden_w.value.strip() else "None"

    return f"""
You are an elite viral short-form content strategist.

Generate EXACTLY 10 viral hooks.

CONTEXT:
Platform: {platform_w.value}
Audience Location: {audience_w.value}
Niche: {niche_w.value}
Clip Type: {clip_type_w.value}
Emotion Target: {emotion_w.value}
Creator Vibe: {vibe_w.value}
Video Length: {length_w.value}
Goal: {goal_w.value}
Notable Individuals: {notable_individuals_w.value}
Forbidden Words: {forbidden}
Reference Hooks:
{ref_lines}


RULES:
- Add notable individuals names to captions, whenever possible
- Each hook must be under 12 words
- Each hook must create curiosity
- Each hook must feel natural, not spammy
- Avoid forbidden words completely
- Do NOT explain anything
- Output ONLY the hooks as a numbered list (1-10)

Generate now:
""".strip()

def generate_hooks(_):
    with output:
        clear_output()
        prompt = build_prompt()

        resp = client.responses.create(
            model="gpt-4.1-mini",
            input=prompt
        )
        raw = resp.output_text.strip()

        # Clean into list (handles slightly messy model output)
        lines = [ln.strip() for ln in raw.split("\n") if ln.strip()]
        hooks = []
        for ln in lines:
            # Remove leading numbering/bullets if present
            ln2 = ln
            if ". " in ln2[:6]:
                ln2 = ln2.split(". ", 1)[1].strip()
            ln2 = ln2.lstrip("-• ").strip()
            if ln2:
                hooks.append(ln2)

        # Ensure exactly 20 display lines
        hooks = hooks[:10]

        print("✅ Generated Hooks:\n")
        for i, h in enumerate(hooks, 1):
            print(f"{i}. {h}")

generate_btn.on_click(generate_hooks)

# Display form
display(
    widgets.VBox([
        platform_w,
        audience_w,
        niche_w,
        clip_type_w,
        emotion_w,
        vibe_w,
        length_w,
        goal_w,
        notable_individuals_w,
        forbidden_w,
        refs_w,
        generate_btn,
        output
    ])
)
```
